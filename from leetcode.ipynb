```python
'''
98. Validate Binary Search Tree
'''
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        def inorder(tree):
            if not tree:
                return []
            if not tree.left and not tree.right:
                return [tree.val]
            return inorder(tree.left) + [tree.val] + inorder(tree.right)
        
        ex_tree = inorder(root)
        for i in range(len(ex_tree)-1):
            if ex_tree[i] >= ex_tree[i + 1]:
                return False
        return True
```

```python
'''
2. Add two numbers
'''
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        num1, num2 = '', ''
        while l1:
            num1 += str(l1.val)
            l1 = l1.next
        while l2:
            num2 += str(l2.val)
            l2 = l2.next
        ans = str(int(num1[::-1]) + int(num2[::-1]))[::-1]
        head = ListNode(val=None)
        cur = head
        for digit in ans:
            cur.next = ListNode(val=int(digit))
            cur = cur.next
        return head.next
```

```python
'''
6. Zigzag Conversion
'''
class Solution:
    def convert(self, s: str, numRows: int) -> str:
        rows = [[] for _ in range(numRows)]
        i, row, descend = 0, 1, False
        while i < len(s):
            rows[row - 1].append(s[i])
            i += 1
            if numRows > 1:
                if row == numRows:
                    descend = True
                if row == 1:
                    descend = False
                if descend:
                    row -= 1
                else: 
                    row += 1
        s = ''
        for row in rows:
            s += ''.join(row)
        return s
```

```python
'''
102. Binary Tree Level Order Traversal
'''
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        queue, ans = [root], [[]]
        cur, nxt = 1, 0
        while queue:
            cur -= 1
            node = queue.pop(0)
            ans[-1].append(node.val)
            if node.left:
                nxt += 1
                queue.append(node.left)
            if node.right:
                nxt += 1
                queue.append(node.right)
            if cur == 0:
                ans.append([])
                cur, nxt = nxt, 0
        ans.pop()
        return ans
```

```python
'''
852. Peak Index in a Mountain Array
'''
class Solution:
    def peakIndexInMountainArray(self, arr: List[int]) -> int:
        l, r = 0, len(arr) - 1
        while l < r-1:
            mid = (l+r)//2
            if arr[mid] > arr[mid-1] and arr[mid] > arr[mid+1]:
                return mid
            elif arr[mid] > arr[mid+1] and arr[mid] < arr[mid-1]:
                r = mid
            elif arr[mid] < arr[mid+1] and arr[mid] > arr[mid-1]:
                l = mid+1
        if arr[l] < arr[r]:
            return r
        else:
            return l
```
